```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#BTCUSD
df0 = pd.read_json('/home/isabelle/Documents/ix-ds-2018/crypto_test/tBTCUSD0.json')

print(df0.shape)

df1 = pd.read_json('/home/isabelle/Documents/ix-ds-2018/crypto_test/tBTCUSD1.json')
df2 = pd.read_json('/home/isabelle/Documents/ix-ds-2018/crypto_test/tBTCUSD2.json')
df3 = pd.read_json('/home/isabelle/Documents/ix-ds-2018/crypto_test/tBTCUSD3.json')
df4 = pd.read_json('/home/isabelle/Documents/ix-ds-2018/crypto_test/tBTCUSD4.json')
df5 = pd.read_json('/home/isabelle/Documents/ix-ds-2018/crypto_test/tBTCUSD5.json')
df_BTCUSD = pd.concat([df0, df1, df2, df3, df4, df5])
df_BTCUSD.columns = ['ID', 'MTS', 'AMOUNT', 'PRICE']
df_BTCUSD.index = range(720)

print(df_BTCUSD.shape)
print(df_BTCUSD.tail())
```

```python
dftime = np.linspace(0,3600,720)
df_BTCUSD['Time'] = dftime
print(df_BTCUSD.head())
```

```python
#normalize price
dfnormalprice = df_BTCUSD['PRICE']/6351.1
df_BTCUSD['Price Index'] = dfnormalprice

#normalise amount
dfnormalamount = df_BTCUSD['AMOUNT']/0.217057
df_BTCUSD['Amount Index'] = dfnormalamount
print(df_BTCUSD.tail())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')


df_BTCUSD.plot(x = 'Time', y = 'Price Index')
df_BTCUSD.plot(x = 'Time', y = 'AMOUNT')
```
